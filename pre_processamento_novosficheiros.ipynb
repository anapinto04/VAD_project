```python
# Aplicar todas as transformações

# Dicionário com todas as colunas e seus respectivos mapeamentos
colunas_mapeamentos = {
    'Cond Aderência': mapeamento_cond_aderencia,
    'Tipos Vias': mapeamento_tipos_vias,
    'Factores Atmosféricos': mapeamento_factores_atmosf,
    'Reg Circulação1': mapeamento_reg_circulacao,
    'Intersecção Vias': mapeamento_interseccao,
    'Localizações': mapeamento_localizacoes,
    'Luminosidade': mapeamento_luminosidade,
    'Marca Via': mapeamento_marca_via,
    'Natureza': mapeamento_natureza,
    'Obras Arte': mapeamento_obras_arte,
    'Obstáculos': mapeamento_obstaculos,
    'Sentidos': mapeamento_sentidos,
    'Sinais': mapeamento_sinais,
    'Sinais Luminosos': mapeamento_sinais_luminosos,
    'Tipo Piso': mapeamento_tipo_piso,
    'Traçado 1': mapeamento_tracado1,
    'Traçado 2': mapeamento_tracado2,
    'Traçado 3': mapeamento_tracado3,
    'Traçado 4': mapeamento_tracado4,
    'Via Trânsito': mapeamento_via_transito
}

# Aplicar transformações para cada coluna
colunas_transformadas = []
colunas_nao_encontradas = []

for nome_coluna, mapeamento in colunas_mapeamentos.items():
    if nome_coluna in df.columns:
        if df[nome_coluna].dtype in ['int64', 'float64', 'int32']:
            df[nome_coluna] = df[nome_coluna].map(mapeamento)
        else:
            df[nome_coluna] = df[nome_coluna].apply(
                lambda x: mapeamento.get(int(x), x) if str(x).isdigit() else x
            )
        colunas_transformadas.append(nome_coluna)
    else:
        colunas_nao_encontradas.append(nome_coluna)

print(f"✓ Transformadas {len(colunas_transformadas)} colunas:")
for col in colunas_transformadas:
    print(f"  - {col}")

if colunas_nao_encontradas:
    print(f"\n⚠ {len(colunas_nao_encontradas)} colunas não encontradas:")
    for col in colunas_nao_encontradas:
        print(f"  - {col}")
```
