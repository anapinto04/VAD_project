```python
import pandas as pd
```

```python
df = pd.read_excel('Tabela_acidentes_30dias_2024_prov.xlsx')
df.head()  
```

```python
import re

# Função para limpar caracteres problemáticos
def limpar_texto(valor):
    if isinstance(valor, str):
        
        valor = valor.replace('\xa0', ' ') 
        valor = valor.replace('\xae', 'ã') 
        valor = valor.replace('Æ', 'ã')
        valor = valor.replace('æ', 'ã')
        valor = valor.replace('‡', 'ç')  
        valor = valor.replace('¡', 'í')
        valor = valor.replace('NÇO DEFINIDO', 'não definido')
        valor = valor.replace('ƒ', 'â')
        valor = valor.replace('äes', 'ões')
        valor = valor.replace('Mˆs', 'Mês')
        valor = valor.replace('aplic vel', 'aplicável')
        valor = valor.replace('Cedˆncia', 'Cedência')
        valor = valor.replace('Obst culos', 'Obstáculos')
        valor = valor.replace('rodovi rias ', 'rodoviárias ')
        valor = valor.replace('Atmosf‚ricos', 'Atmosféricos')
       
        valor = ' '.join(valor.split())
        return valor
    return valor

df.columns = [limpar_texto(col) for col in df.columns]

for column in df.columns:
    if df[column].dtype == 'object':
        df[column] = df[column].apply(limpar_texto)


colunas_detalhadas = ['Cod Via','vitimas mortais 30 dias','Traçado 1','Traçado 2', "Traçado 3", "Traçado 4", "Via Trânsito"]  # Substitua pelos nomes das colunas desejadas

for i, column in enumerate(df.columns):
    print(f"\n{'='*50}")
    print(f"Coluna: {column}")
    print(f"Total de registros: {len(df[column])}")
    print(f"Valores nulos: {df[column].isna().sum()}")
    print(f"Valores únicos: {df[column].nunique()}")
    

    if column in colunas_detalhadas:
        print(f"\nValores únicos:")
        unique_values = df[column].dropna().unique()
        for value in unique_values:
            count = (df[column] == value).sum()
            print(f"  - {value} ({count})")
        if df[column].isna().sum() > 0:
            print(f"  - [NÃO DEFINIDO] ({df[column].isna().sum()})")
```

```python
# Salvar o dataframe limpo em CSV
df.to_csv('Tabela_acidentes_30dias_2024_limpo.csv', index=False, encoding='utf-8')
```
