```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_2018=pd.read_excel('Datasets_Limpos\\Tabela_acidentes_2018_limpo.xlsx')
df_2019=pd.read_excel('Datasets_Limpos\\Tabela_acidentes_2019_limpo.xlsx')
df_2020=pd.read_excel('Datasets_Limpos\\Tabela_acidentes_2020_limpo.xlsx')
df_2021=pd.read_excel('Datasets_Limpos\\Tabela_acidentes_2021_limpo.xlsx')
df_2022=pd.read_excel('Datasets_Limpos\\Tabela_acidentes_2022_limpo.xlsx')
df_2023=pd.read_excel('Datasets_Limpos\\Tabela_acidentes_2023_limpo.xlsx')
df_2024=pd.read_csv('Datasets_Limpos\\Tabela_acidentes_30dias_2024_limpo.csv')
```

```python
import pandas as pd

def descreve_variaveis(df):

    variaveis_numericas = df.select_dtypes(include="number").columns.tolist()
    variaveis_categoricas = df.select_dtypes(exclude="number").columns.tolist()
    return {
        "variaveis_numericas": variaveis_numericas,
        "variaveis_categoricas": variaveis_categoricas,
        "n_num": len(variaveis_numericas),
        "n_cat": len(variaveis_categoricas),
    }

# exemplo de uso: agrupe seus dataframes num dicionário para iterar

dados = {
    "2018": df_2018,
    "2019": df_2019,
    "2020": df_2020,
    "2021": df_2021,
    "2022": df_2022,
    "2023": df_2023,
    "2024": df_2024,
}

for ano, df in dados.items():
    info = descreve_variaveis(df)
    print(f"\n{ano} – numéricas: {info['n_num']}, categóricas: {info['n_cat']}")
    print("  numéricas:", info["variaveis_numericas"])
    print("  categóricas:", info["variaveis_categoricas"])
```

```python
# plot histograms for every numeric variable in each year's dataframe
dados = {
    "2018": df_2018,
    "2019": df_2019,
    "2020": df_2020,
    "2021": df_2021,
    "2022": df_2022,
    "2023": df_2023,
    "2024": df_2024,
}

for ano, df in dados.items():
    variaveis_numericas = df.select_dtypes(include="number").columns.tolist()
    for column in variaveis_numericas:
        plt.figure(figsize=(4, 4))
        df[column].hist()
        plt.xlabel(column)
        plt.ylabel('Frequência')
        plt.title(f'Histograma de {column} ({ano})')
        plt.show()
```
